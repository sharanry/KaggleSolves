# Visualisations for Digit Recognision





```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import matplotlib.cm as cm
%matplotlib inline  
from sklearn.decomposition import PCA


# image number to output
IMAGE_TO_DISPLAY = 20212
```

## Accessing the data.

```python
data = pd.read_csv('train.csv')
data_t = pd.read_csv('test.csv')

print('data({0[0]},{0[1]})'.format(data.shape))
print (data.head())
```

## Preprocessing

```python
images = data.iloc[:,1:].values
images_t = data_t.iloc[:,:].values 
images = images.astype(np.float)
images_t = images_t.astype(np.float)
images = np.multiply(images, 1.0 / 255.0)
images_t = np.multiply(images_t, 1.0 / 255.0)
print('images({0[0]},{0[1]})'.format(images.shape))
print('images_t({0[0]},{0[1]})'.format(images_t.shape))
```

```python
# display image
def display(img):
    
    # (784) => (28,28)
    one_image = img.reshape(28,28)
    
    plt.axis('off')
    plt.imshow(one_image, cmap=cm.binary)

# output image     
display(images[10])
```

```python
X = images
y = data.iloc[:,0].values
print(y.shape[0])
```

## Reducing dimention to two for visualisations

```python
pca2 = PCA(n_components=2,svd_solver ='auto', whiten = True)
pca2.fit(X)
X_ = pca2.transform(X)
print(X_.shape, pca2.noise_variance_)
```

## 2D scatter plot

```python
plt.scatter(X_[:,0], X_[:,1], c=y,s=0.01)
```

## 3D scatter plot

```python
pca3 = PCA(n_components=3,svd_solver ='auto', whiten = True)
pca3.fit(X)
X_3 = pca3.transform(X)
print(X_3.shape, pca2.noise_variance_)


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_3[:,0], X_3[:,1],X_3[:,2], c=y, zdir ='y', depthshade=False, s=0.01)
ax.legend()
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')

ax1.scatter(X_3[:,0], X_3[:,1],y, c=y, zdir ='z', depthshade=False, s=0.01)
ax1.legend()
```

```python
pca1 = PCA(n_components=1,svd_solver ='auto', whiten = True)
pca1.fit(X)
X_ = pca1.transform(X)
print(X_.shape, pca1.noise_variance_)
plt.scatter(X_[:,0], y, c=y,s=0.01)
```
